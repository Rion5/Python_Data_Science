## DataFrames from dict (1)

Keys of the dictionary data are used as column labels

```python
data = { 
        'weekday' : ['Mon', 'Mon', 'Tues', 'Tues'],
        'city' : ['Chicago', 'Indianapolis', 'Chicago', 'Indianapolis'],
        'visitors': [230, 150, 128, 98]
       }
df1 = pd.DataFrame(data)
df1.head()
```

## DataFrames from dict (2)

```python
weekdays = ['Mon', 'Mon', 'Tues', 'Tues']
cities = ['Chicago', 'Indianapolis', 'Chicago', 'Indianapolis']
visitors = [230, 150, 128, 98]
list_labels = ['weekdays', 'city', 'visitors']
list_cols = [weekdays, cities, visitors] # Note: list_cols is a list of lists
# Using Python's list() and zip() functions, zip 2 lists together into a list of tuples called zipped
zipped = list(zip(list_labels, list_cols))
zipped
```

```python
data = dict(zipped)
df2 = pd.DataFrame(data)
df2.head()
```

## Broadcasting

Broadcasting is a convenient technique in NumPy & pandas

Using the previous dataframe (df2) we can easily create a new column, 'fees' on the fly.

```python
df2['fees'] = 0 # Broadcasts to entire column
df2.head()
```
