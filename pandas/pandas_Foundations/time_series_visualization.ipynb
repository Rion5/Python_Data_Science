# Time Series Visualization

* Line Types
* Plot Types
* Subplots

**Microsoft Stock Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as wb

MSFT = wb.DataReader('MSFT', data_source='yahoo', start='2010-01-01')
MSFT.tail()
```

## Pandas Plot

```python
MSFT['Close'].plot()
plt.show()
```

### Labels and Titles

```python
MSFT['Close'].plot(title='Microsoft Stock Data from 2010 - Present')
plt.ylabel('Closing Price (US Dollars)')
plt.show()
```

#### 1st Week of April in 2012

```python
MSFT.loc['2012-4-1' : '2012-4-7', 'Close'].plot(title='MSFT Stock - 1st Week in April 2012')
plt.ylabel('Closing Price (US Dollars)')
plt.show()
```

### Plot Styles

* Style Format String
    * color (k = black)
    * marker: (. = dot)
    * line type(- = solid)
  
| Color  | Marker   | Line     |
|--------|----------|----------|
|b: blue |o: circle |: : dotted|
|g: green|* : Star  |-: dashed |
|r: red  |s: square |   |
|c: cyan |+: plus   |   |

```python
MSFT.loc['2012-4', 'Close'].plot(title='MSFT Weekly Stock Closing Price - April 2012', style='k.-')
plt.ylabel('Closing Price (US Dollars)')
plt.show()
```

#### Area Plot

```python
MSFT['Close'].plot(kind='area', title='Microsoft Closing Price from 2010 - Present')
plt.ylabel('Closing Price (US Dollars)')
plt.show()
```

#### Multiple Columns

subplots = **True** will make a separate plot with separate y-axis for each column

```python
MSFT.loc['2012', ['Close', 'Volume']].plot(title='Microsoft Stock 2012', subplots=True)
plt.show()
```
